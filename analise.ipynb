```python
df_order = df_bank.sort_values(by='balance', ascending=True)
```

```python
df_order['type'] = pd.qcut(df_order['balance'], q=[0,0.8,0.9,1], labels=['Normal', 'Gold', 'Premium'])

df_order
```

```python
df_order[(df_order['type'] == 'Premium')].count()
```

```python
# 3) Seu chefe também quer saber se existe diferença nos percentuais vistos no exercício (1) para o grupo Premium. 
# Verifique como a tabela cruzada da população se diferencia do grupo Premium. 
# Existe diferença? Qual sua hipótese para que exista essa diferença?

pd.crosstab([df_bank['marital'], df_bank[(df_bank['job'] == 'entrepreneur')]['job']], 
            df_bank['type'], margins=False).apply(lambda r: (r/r.sum()) * 100, axis=0)

# Existe uma diferença entre o percentual geral de divorciados que são empreendedores no geral e 
# divorciados que são empreendedores
# dentro da categoria Premium
```

```python
# 4) Na sua opinião e de parte da diretoria, 
# há 3 fatores que determinam se a pessoa vai aceitar a proposta do seu banco (y): 
# o saldo (balance), se ela já deixou de pagar um empréstimo (default) e se ela tem um empréstimo pessoal (loan). 
# Para avaliar essas 3 características junto com o aceite (y), 
# você precisará criar uma tabela dinâmica, cujo valor observado seja a média do saldo. 
# Avalie se existe diferença no saldo de quem aceita a proposta e tem empréstimo vs quem não aceita e não tem empréstimo, 
# e o mesmo para o default. Em outras palavras, crie uma tabela dinâmica cruzando loan e default, 
# mas também segmentando por aceite ou não da proposta. O saldo de quem aceita e 
# tem empréstimo é maior ou menor do que o de quem não aceita? 
# Por qual motivo você acha que isso acontece? Crie a sua hipótese para justificar o resultado achado!


df_bank.pivot_table(index=['y', 'default'], columns='loan', values='balance', aggfunc=np.mean)
```
